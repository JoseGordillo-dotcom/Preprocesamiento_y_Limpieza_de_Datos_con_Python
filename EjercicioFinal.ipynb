```python
import pandas as pd
```

```python
df = pd.read_csv("datos/titanic.csv")
df
```

```python
df['age'] = round(df.age.fillna(df.age.mean()), 1)
df
```

```python
df['embarked'] = df.embarked.fillna('S')
df
```

```python
df['sex'] = df['sex'].replace({"female":1, "male":0})
df
```

```python
df['embarked'] = df['embarked'].replace({"S":0, "C":1, "Q":2})
df
```

```python
child_rango = [0, 16, 99]
child_etiqueta = [1, 0]
df['child'] = pd.cut(df['age'], bins=child_rango, labels=child_etiqueta)
df
```

```python
df.drop(['name', 'ticket', 'cabin'], axis=1, inplace = True)
df
```

```python
df.duplicated()
```

```python
df.drop_duplicates(keep='first', inplace = True)
df
```

```python
df['sex'].unique()
```

```python
df.info()
```

```python
df['boat'] = df['boat'].str.strip()
df
```

```python
df['home.dest'] = df['home.dest'].str.strip()
df
```

```python
df.info()
```

```python
#Escribir JSON orientado a registros
df.to_json('datos/df/JRJGN_titanic_clean.json', indent = 2,
           orient="records")
```
